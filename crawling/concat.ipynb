#### google scholar 데이터 중복 제거 및 존재 하지 않는 초록까지 제거 

```python
import pandas as pd
```

```python
data = pd.read_excel('google scholar data.xlsx')
```

```python
data = data.drop(['Unnamed: 0'],axis = 1)
```

```python
data= data.rename(columns = {'논문 제목':'title','저널':'journal','연도':'year','초록':'abstracts'})
```

```python
data = data.dropna()
```

```python
data = data[data['abstracts'] != 'No abstract available']
```

```python
data['journal'].value_counts()
```

```python
data = data[data['journal'] == '한국운동역학회지,']
```

```python
data['journal'] = data['journal'].str.replace(',', '', regex=False)
```

```python
data[data.duplicated(keep=False)]['title'].value_counts()
```

```python
data = data.drop_duplicates(['title'])
```

```python
data[data.duplicated(keep=False)]['abstracts'].value_counts()
```

```python
data = data.drop_duplicates(['abstracts'])
```

```python
data
```

#### 데이터 합치기 (google + riss)

```python
dt = pd.concat([data,riss]) #2325
```

```python
# dt_.to_excel('중복확인.xlsx')
```

* google, riss concat 하였을때 제목 중복 데이터 개수

* 중복 제거 후 남길 데이터 지정 -> Keep = False 전부제거, Keep = first 첫번째 행 남기고 제거, keep = last 맨 마지막 행 남기고 제거

```python
dt_[dt_.duplicated()] #180 중복된 데이터
```

```python
dt_.drop_duplicates(inplace = True) #중복된 180개 데이터 제거
```

```python
dt_
```

```python
dt_ = dt_.sort_values(['year'],ascending = True)
```

```python
dt_ = dt_.reset_index()
```

```python
dt_ = dt_.drop(['index'],axis = 1)
```

#### 중복 되지 않은 데이터, 즉 google에는 없고 riss에는 있는 데이터 

```python
# 두 데이터프레임 병합 (indicator=True로 병합 상태 표시)
merged = pd.merge(data, riss, how='outer', indicator=True)

# data에는 없고 riss에는 있는 데이터 선택
riss_only = merged[merged['_merge'] == 'right_only']

# 결과 출력
riss_only = riss_only.sort_values(['year'],ascending = True)
riss_only = riss_only.reset_index()
riss_only = riss_only.drop(['index'],axis = 1)
riss_only
```
