### Google Scholar 논문 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
import pandas as pd
```

#### 1992 - 2000년도

```python
# 데이터 저장할 리스트
data = []

# WebDriver 설정
driver = webdriver.Chrome()

# Google Scholar 페이지로 이동 및 데이터 수집
keyword = "한국운동역학회"
for start in range(0, 131, 10):
    print(f"Processing page starting with {start}")
    url = f"https://scholar.google.co.kr/scholar?start={start}&q={keyword}&hl=ko&as_sdt=8%2C5&as_ylo=1992&as_yhi=2000"
    
    driver.get(url)
    time.sleep(3)  # 페이지 로딩 대기

    # 페이지 소스 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # 논문 제목 링크 추출
    paper_tags = soup.find_all("div", class_="gs_ri")

    for paper in paper_tags:
        title_tag = paper.find("h3", class_="gs_rt")
        if title_tag and title_tag.a:
            title_text = title_tag.a.text
            title_link = title_tag.a['href']

            # 연도 및 저널 정보 추출
            try:
                year_info = paper.find("div", class_="gs_a").get_text()
                year = year_info.split()[-3].strip()
                journal = year_info.split()[-4].strip()
            except AttributeError:
                year = "Unknown"
                journal = "Unknown"
            
            # 논문 제목 링크 클릭 후 초록 가져오기
            driver.get(title_link)
            time.sleep(3)  # 페이지 로딩 대기

            # 초록 가져오기
            abs_html = driver.page_source
            abs_soup = BeautifulSoup(abs_html, 'html.parser')
            abstract = ""
            
            if abs_soup.select_one('.abstractTxt'):
                abstract = abs_soup.select_one('.abstractTxt').text.strip()
            elif abs_soup.select_one('.g-color-gray-dark-v2.text-justify'):
                abstract = abs_soup.select_one('.g-color-gray-dark-v2.text-justify').text.strip()
            elif abs_soup.select_one('.first'):
                abstract = abs_soup.select_one('.first').text.strip()
            elif abs_soup.select_one('.AbstractText'):
                abstract = abs_soup.select_one('.AbstractText').text.strip()
            elif abs_soup.select_one('.wrap_part_body'):
                abstract = abs_soup.select_one('.wrap_part_body').text.strip()
            else:
                abstract = "No abstract available"
            
            data.append({"논문 제목": title_text, "저널": journal, "연도": year, "초록": abstract})

            driver.back()
            time.sleep(3)  # 페이지 로딩 대기

driver.quit()

# 데이터프레임으로 변환 후 엑셀로 저장
df1 = pd.DataFrame(data)
```
